```python
def altnullhist(scores, lfdr, nbins=100, title_text="altnullhist", score_name="pvalue", alt_ontopof_null=True):
    min_score, max_score = min(scores), max(scores)
    score_bins = np.linspace(min_score, max_score, nbins)
    bin_cnt = np.zeros(nbins)
    sumlfdr = np.zeros(nbins)
    for i, score in enumerate(scores):
        bin_idx = nbins-1 if score == max_score else int((score - min_score) / (max_score - min_score) * nbins)
        bin_cnt[bin_idx] += 1
        sumlfdr[bin_idx] += lfdr[i]
        
    
    avlfdr = sumlfdr / bin_cnt
    full_den = bin_cnt / (np.sum(bin_cnt) * (max_score - min_score) / nbins)
    null_den = avlfdr * full_den
    alt_den = full_den - null_den

    df = pd.DataFrame({
        score_name: score_bins,
        "null_den": null_den,
        "alt_den": alt_den,
    })
    
    if alt_ontopof_null:
        fig = px.bar(df, x=score_name, y=["alt_den", "null_den"])
        fig.update_layout(title_text=title_text, width=800, height=500)
        fig.show()
    else:        
        fig = px.bar(df, x=score_name, y=["null_den", "alt_den"])
        fig.update_layout(title_text=title_text, width=800, height=500)
        fig.show()

altnullhist(zscores, loc_fdr, score_name="zscore")
```
